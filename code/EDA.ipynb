```python
# here we can see the quantity, price per unit, total spent can be float/int & transaction date can be date
data['Quantity'] = pd.to_numeric(data['Quantity'], errors = 'coerce')
data['Price Per Unit'] = pd.to_numeric(data['Price Per Unit'], errors = 'coerce')
data['Total Spent'] = pd.to_numeric(data['Total Spent'], errors = 'coerce')
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'], errors = 'coerce')
```

```python
# we do not have any duplicates in the data, but if we have any duplicates we can remove in this way
data = data.drop_duplicates()
```

```python
data.isnull().sum()
```

```python
data.groupby("Item")["Price Per Unit"].unique()
# this returns unique values in the both the columns
```

```python
price_map = data.groupby("Item")["Price Per Unit"].mean()
print(price_map)
# in this way we create a price map
```

```python
data['Price Per Unit'] = data['Price Per Unit'].fillna(data['Item'].map(price_map))
```

```python
data.isnull().sum()
```

```python
data['Payment Method'] = data['Payment Method'].fillna(data['Payment Method'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['Location'].unique()
```

```python
data["Location"].value_counts(dropna=False)
```

```python
loc_map = data.groupby("Payment Method")["Location"].agg(lambda x: x.mode()[0])
print(loc_map)
```

```python

data["Location"] = data["Location"].fillna(data["Payment Method"].map(loc_map))
```

```python
data["Transaction Date"].fillna(data["Transaction Date"].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
price_map = data.groupby("Item")["Price Per Unit"].median()
print(price_map)
```

```python
reverse_map = (price_map.reset_index().groupby("Price Per Unit")["Item"].first())

print(reverse_map)
```

```python
data["Item"] = data["Item"].fillna(data["Price Per Unit"].map(reverse_map))
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=["Item", "Price Per Unit"])
```

```python
data.isnull().sum()
```

```python
data['Total Spent'] = data['Total Spent'].fillna(data['Quantity']*data['Price Per Unit'])
```

```python
data.isnull().sum()
```

```python
data['Quantity'] = data['Quantity'].fillna(data['Total Spent'] / data['Price Per Unit'])
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset = ['Quantity','Total Spent'])
```
